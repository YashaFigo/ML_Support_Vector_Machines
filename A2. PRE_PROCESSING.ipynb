### HAND-SCALING THE DATA

```python
data[:5]
```

><b>How it works:   

><b>The data array has two columns, column1 will be the x axis and column2 the y axis
The mean and standard deviation are computed for each column.  

><b>For each column:     

><b>scaled value = original_value - column_mean / column_std


```python
hand_scaled_data = []

mean = np.mean(data)
std = np.std(data)

col_1 = [x[0] for x in data]
col_2 = [x[1] for x in data]

minimum_1 = min(col_1)
maximum_1 = max(col_1)
mean_1 = np.mean(col_1)
std_1 = np.std(col_1)

minimum_2 = min(col_2)
maximum_2 = max(col_2)
mean_2 = np.mean(col_2)
std_2 = np.std(col_2)

for x in data:
    a = x[0]
    b = x[1]
    
    a_scaled = (a - mean_1 )/( std_1 )
    b_scaled = (b - mean_2 )/( std_2 )
    
    hand_scaled_data.append([a_scaled,b_scaled])

hand_scaled_data = array(hand_scaled_data)

print(scaled_data[:10])
print('')
print(hand_scaled_data[:10])

plt.scatter([x[0] for x in scaled_data],[x[1] for x in scaled_data])
plt.show()

plt.scatter([x[0] for x in hand_scaled_data],[x[1] for x in hand_scaled_data])
plt.show()
```
